```python
import random as r
import matplotlib.pyplot as plt
```

```python
n = 1000000
```

```python
p = 0
l = []
for i in range(1, n+1):
    x = r.uniform(-1, 1)
    y = r.uniform(-1, 1)
    if x**2+y**2<=1:
        p+=1
    l.append((p/i)*4)
```

```python
plt.plot(l)
plt.ylim(2.75, 3.55)
plt.show()
```
