```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
deg = np.arange(0, 361, 1)
x = np.cos(np.radians(deg))
y = np.sin(np.radians(deg))
```

```python
plt.plot(x, y)
```

```python
for i in range(0, 10, 2):
    plt.plot(x*i, y*i, color="blue")
    plt.plot(x*(i+0.5), y*(i+0.5), color="orange", linestyle="dotted")
    plt.plot(x*(i+1), y*(i+1), color="green", linestyle="dashed")
    plt.plot(x*(i+1.5), y*(i+1.5), color="red", linestyle="dashdot")
plt.show()
```
